```python
from sklearn import datasets,svm,metrics
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
from skimage import data,color,transform
```

```python
digits = datasets.load_digits()
```

```python
images = digits.images
print(type(images))#images are already flattened
print(images.shape)
```

```python
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
print(data.shape)
print(type(data))
```

```python
for i in range(10):
    plt.matshow(transform.resize(digits.images[i],(32,32)))
```

**When we reszie our sample images from 8X8 to 64X64**

```python
def Img64X64(gamma_user):
    resized_image64=[]
    for i in digits.images:
        resized_image64.append(transform.rescale(i,8,anti_aliasing=False))
    resized_image64 = np.array(resized_image64)
    data = resized_image64.reshape((n_samples,-1))
    classifier = svm.SVC(gamma=gamma_user)
    X_train64,X_test64,Y_train64,Y_test64 = train_test_split(data,digits.target,train_size=0.75,shuffle=False)
    classifier.fit(X_train64,Y_train64)
    pred64_train = classifier.predict(X_train64)
    acc64_train = metrics.accuracy_score(Y_train64,pred64_train)
    pred64 = classifier.predict(X_test64)
    acc64 = metrics.accuracy_score(Y_test64,pred64)
    return acc64_train*100,acc64*100
```

**When we reszie our sample images from 8X8 to 32X32**

```python
def Img32X32(gamma_user):
    resized_image32=[]
    for i in digits.images:
        resized_image32.append(transform.resize(i,(32,32),anti_aliasing=False))
    resized_image32 = np.array(resized_image32)
    data = resized_image32.reshape((n_samples,-1))
    classifier32 = svm.SVC(gamma=0.001)
    X_train32,X_test32,Y_train32,Y_test32 = train_test_split(data,digits.target,train_size=0.75,shuffle=False)
    classifier32.fit(X_train32,Y_train32)
    pred32 = classifier32.predict(X_test32)
    acc32 = metrics.accuracy_score(Y_test32,pred32)
    return acc32*100
```

**When we reszie our sample images from 8X8 to 4X4**

```python
def Img4X4(gamma_user):
    resized_image4=[]
    for i in digits.images:
        resized_image4.append(transform.resize(i,(4,4),anti_aliasing=False))
    resized_image4 = np.array(resized_image4)
    data = resized_image4.reshape((n_samples,-1))
    classifier4 = svm.SVC(gamma=0.001)
    X_train4,X_test4,Y_train4,Y_test4 = train_test_split(data,digits.target,train_size=0.75,shuffle=False)
    classifier4.fit(X_train4,Y_train4)
    pred4 = classifier4.predict(X_test4)
    acc4 = metrics.accuracy_score(Y_test4,pred4)
    return acc4*100
```

```python
print("Accuracy for 64X64 images on 75:25 Split and gamma 0.0001 on train set and test set :- ",Img64X64(0.0001))
print("Accuracy for 64X64 images on 75:25 Split and gamma 1 on train set and test set :- ",Img64X64(1))
print("Accuracy for 64X64 images on 75:25 Split and gamma 0.5 on train set and test set :- ",Img64X64(0.5))
```

```python

print("Accuracy for 32X32 images on 75:25 Split and gamma 0.0001 :- ",Img32X32(0.0001))
print("Accuracy for 4X4 images on 75:25 Split and gamma 0.0001 :- ",Img4X4(0.0001))
```

```python
print("Accuracy for 64X64 images on 75:25 Split and gamma 0.01 :- ",Img64X64(0.01))
print("Accuracy for 32X32 images on 75:25 Split and gamma 0.01 :- ",Img32X32(0.01))
print("Accuracy for 4X4 images on 75:25 Split and gamma 0.01 :- ",Img4X4(0.01))
```
