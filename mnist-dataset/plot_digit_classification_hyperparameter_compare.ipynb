```python
from sklearn import datasets,svm,metrics
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
from skimage import data,color,transform
import seaborn as sns
import pandas as pd
```

```python
digits = datasets.load_digits()
```

```python
images = digits.images
print(type(images))#images are already flattened
print(images.shape)
```

```python
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
print(data.shape)
print(type(data))
```

**When we reszie our sample images from 8X8 to 64X64**

```python
def Img64X64(gamma_user):
    resized_image64=[]
    for i in digits.images:
        resized_image64.append(transform.rescale(i,8,anti_aliasing=False))
    resized_image64 = np.array(resized_image64)
    data = resized_image64.reshape((n_samples,-1))
    classifier = svm.SVC(gamma=gamma_user)
    X_train64,X_test64,Y_train64,Y_test64 = train_test_split(data,digits.target,test_size=0.3,shuffle=False)
    X_test64,X_val64,Y_test64,Y_val64 = train_test_split(X_test64,Y_test64,test_size=0.5,shuffle=False)
    classifier.fit(X_train64,Y_train64)
    pred64_val = classifier.predict(X_val64)
    acc64_val = metrics.accuracy_score(Y_val64,pred64_val)
    pred64 = classifier.predict(X_test64)
    acc64 = metrics.accuracy_score(Y_test64,pred64)
    return acc64_val*100,acc64*100
```

**When we reszie our sample images from 8X8 to 32X32**

```python
def Img32X32(gamma_user):
    resized_image32=[]
    for i in digits.images:
        resized_image32.append(transform.resize(i,(32,32),anti_aliasing=False))
    resized_image32 = np.array(resized_image32)
    data = resized_image32.reshape((n_samples,-1))
    classifier32 = svm.SVC(gamma=gamma_user)
    X_train32,X_test32,Y_train32,Y_test32 = train_test_split(data,digits.target,train_size=0.7,shuffle=False)
    X_test32,X_val32,Y_test32,Y_val32 = train_test_split(X_test32,Y_test32,test_size=0.5,shuffle=False)
    classifier32.fit(X_train32,Y_train32)
    pred32_val = classifier32.predict(X_val32)
    acc32_val = metrics.accuracy_score(Y_val32,pred32_val)
    pred32 = classifier32.predict(X_test32)
    acc32 = metrics.accuracy_score(Y_test32,pred32)
    return acc32_val*100,acc32*100
```
